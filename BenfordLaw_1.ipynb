```python
#Merge_Municipios_Dian.csv

# Lectura del Dataframe
# Data frame reading
table_file = r"Merge_Municipios_Dian.csv"
df_table = pd.read_csv(table_file)
# Define las variables de agrupación temporal, el atributo de membresía y la variable numérica a analizar 
# Defines the temporal grouping variables, the membership attribute and the numeric variable to analyze
field_id = 'Código Actividad Económica Principal'
var_study = 'Total Ingresos Netos (459)'
time_var = 'Año Gravable'

print(df_table.columns)
print(df_table.shape)
df_table.head()
```

## 1.2 Análisis general de la variable de estudio por año
## 1.2 General analysis of the study variable by year

```python
## análisis de la variable de interés agrupada por la variable temporal 'Año gravable'
## analysis of the interest variable grouped by the time variable 'Taxable year'

df = df_table
for anio in df[time_var].unique():
    bl = benfordslaw(verbose = 0)
    Iloc = df[time_var] == anio
    X = df[var_study].loc[Iloc].values
    bl.fit(X)
    bl.plot(title=str(anio))
    print(X.shape[0])
```

## 1.3 Results for a particular economic activity

```python
# Resultados particulares para una actividad económica principal
# Este fragmento de código permite el análisis particular de uno de los registros dentro de campo de membresía (field_id)

# Particular results for a principal economic activity
# This code snippet allows the particular analysis of one of the records within the membership field (field_id)

df = df_table[df_table[field_id] == 5621]
for anio in df[time_var].unique():
    bl = benfordslaw(verbose = 0)
    Iloc = df[df[time_var] == anio]
    X = Iloc[var_study]
   # X = df['IMPUESTO_DECLARADO'].loc[Iloc].values
    bl.fit(X)
    bl.plot(title=str(anio))
    print(Iloc.shape[0])
```

# 2. Resultados del análisis de Benford para el Dataframe
# 2. Benford analysis results for the Dataframe

```python
'''Se efectúa un análisis de la ley de Benford para cada subconjunto de los datos correspondiente a las agrupaciones
temporales y de membresía. El resultado es un nuevo dataframe con la suma de valores de la variable de interés, el
indicador de anomalía y la cantidad de registros tenidos en cuenta para cada test.

An analysis of Benford's law is performed for each subset of the data corresponding to the clusters
temporary and membership. The result is a new dataframe with the sum of the values of the variable of interest, the
anomaly indicator and the number of records taken into account for each test.
'''

df_table_benford = []
for id in df_table[field_id].unique():
    df1 = df_table[df_table[field_id] == id]
    for time in df1[time_var].unique():
        bl = benfordslaw(verbose = 0)
        Iloc = df1[df1[time_var] == time][var_study]
        Iloc2 = df1[df1[time_var] == time]['NoOmiso']
        bl.fit(Iloc)
        if np.isnan(bl.results['P']):
            flag = 'nan'
        elif bl.results['P']<= 0.05:
            flag = 'Anomaly'
        else: 
            flag = 'No anomaly'  
            
        #print(id,time,Iloc.sum(),flag,Iloc.shape[0])
        df_table_benford.append({field_id: id,time_var: time,var_study+'sum': Iloc.sum(), 'benfords': flag,'benfords_count': Iloc.shape[0], 'NoOmiso_sum':Iloc2.sum()})


df_table_benford = pd.DataFrame(df_table_benford)
df_table_benford.head(10)
```

## 2.1 Subset dataframe inconclusive

```python
# Estadisticas para el grupo de sesultados no concluyentes
# Statistics for the group of inconclusive results

nan = df_table_benford[df_table_benford.benfords == 'nan']
sns.distplot(np.log(nan.benfords_count))
print(' min: ',nan.benfords_count.min(),'\n max: ',nan.benfords_count.max(),'\n mean: ',nan.benfords_count.mean())
print(' quantils: \n',nan.benfords_count.quantile([0.25,0.5,0.75,0.95]))
```

## 2.2 Subset dataframe sin anomalias
## 2.2 Subset dataframe without anomaly

```python
no_anomaly = df_table_benford[df_table_benford.benfords == 'No anomaly']
sns.distplot(np.log(no_anomaly.benfords_count))
print(' min: ',no_anomaly.benfords_count.min(),'\n max: ',no_anomaly.benfords_count.max(),'\n mean: ',no_anomaly.benfords_count.mean())
print(' quantils: \n',no_anomaly.benfords_count.quantile([0.25,0.5,0.75,0.95]))
```
