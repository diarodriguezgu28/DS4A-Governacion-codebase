## 1. Arhivos de las camaras de Comercio de los municipios del Valle del Cauca
## 1. Archives of the Chambers of Commerce of the municipalities of Valle del Cauca

```python

# CC Buenaventura
df_c_c_btura_A_M = pd.read_excel(r"c.c. btura A M.xls")
df_Buenaventura = df_c_c_btura_A_M.dropna(how="all",axis=1)

# CC Cartago
df_CARTAGOAM = pd.read_excel(r"CARTAGO A M.xlsx")
df_Cartago = df_CARTAGOAM.dropna(how="all",axis=1)
df_Cartago = df_Cartago.drop(['CIIU-2', 'CIIU-3', 'CIIU-4'], axis=1)

# CC Cartago PyG
df_CCCARTAGOCONPYGAM = pd.read_excel(r"CC CARTAGO CON PYG A M.xlsx")
df_Cartago_pyg = df_CCCARTAGOCONPYGAM.dropna(how="all",axis=1)

# CC Buga
df_CCBUGAAM = pd.read_excel(r"CC BUGA A M.xlsx")
df_Buga = df_CCBUGAAM.dropna(how="all",axis=1)

# CC Cali
df_ccCaliAM = pd.read_excel(r"cc Cali AM.xlsx")
df_Cali = df_ccCaliAM.dropna(how="all",axis=1)

#df_Cali2 = pd.read_excel(r"cc Cali AM.xlsx", 'EstSucAge')
#df_Cali2 = df_Cali2.dropna(how="all",axis=1)

# CC Palmira
df_ccpalmiraAM = pd.read_excel(r"cc palmira A M.xlsx", skiprows=1)
df_Palmira = df_ccpalmiraAM.dropna(how="all",axis=1)

# CC Sevilla
df_CCSEVILLAAM = pd.read_excel(r"CC SEVILLA A M.xlsx")
df_Sevilla = df_CCSEVILLAAM.dropna(how="all",axis=1)

# CC Sevilla PyG
df_sevillaconAMconpyg = pd.read_excel(r"sevilla con A M  con pyg.xlsx")
df_Sevilla_pyg = df_sevillaconAMconpyg.dropna(how="all",axis=1)

# CC Tulua
df_cctuluaAM = pd.read_excel(r"cc tulua A M.xlsx")
df_Tulua = df_cctuluaAM.dropna(how="all",axis=1)
```

## 2. Cleaning Actividades económicas y estampillas
## 2. Cleaning Economic activities and stamps

```python
#Contribucion al deporte
df_contribucionAM = pd.read_excel(r"contribucion  A M.xlsx")
df_Contrib_deporte = df_contribucionAM.dropna(how="all",axis=1)

#Licores
df_DECLARACIONNaleImportado2015al2019AM = pd.read_excel(r"DECLARACION Nal e Importado 2015 al 2019  AM.xlsx", skiprows=5)
df_Licores = df_DECLARACIONNaleImportado2015al2019AM.dropna(how="all",axis=1)

#Deguello
df_deguelloAM= pd.read_excel(r"deguello  A M.xlsx")
df_Deguello = df_deguelloAM.dropna(how="all",axis=1)

#Formato 240
df_form240AM = pd.read_excel(r"form 240 AM.xlsx")
df_240 = df_form240AM.dropna(how="all",axis=1)

#Formato 110
df_formato110AM = pd.read_excel(r"formato 110 AM.xlsx")
df_110 = df_formato110AM.dropna(how="all",axis=1)

#Formato 210
df_formato210AM = pd.read_excel(r"formato 210 A M.xlsx")
df_210  = df_formato210AM.dropna(how="all",axis=1)

#Formato 210
df_formato230AM = pd.read_excel(r"formato 230 AM.xlsx")
df_230 = df_formato230AM.dropna(how="all",axis=1)

#Loterias
df_loterias_A_M = pd.read_excel(r"loterias  A M.xlsx")
df_Loterias = df_loterias_A_M.dropna(how="all",axis=1)

#Relacion de Ventas
df_RELACIONDEVEntasyPD2015AL2019am = pd.read_excel(r"RELACION DE VEntas y PD 2015 AL 2019 a m.xlsx")
df_Relacion_ventas = df_RELACIONDEVEntasyPD2015AL2019am.dropna(how="all",axis=1)

#Sobretasa Gasolina
df_sobretasaalagasolinaAM= pd.read_excel(r"sobretasa a la gasolina  A M.xlsx")
df_Gasolina = df_sobretasaalagasolinaAM.dropna(how="all",axis=1)

#Ventas
df_ventas300AM = pd.read_excel(r"ventas 300 A M.xlsx")
df_Ventas = df_ventas300AM.dropna(how="all",axis=1)
```

## 3. Homogenización de los nombres de columnas
## 3. Homogenize Column Names

```python
Buenaventura = df_Buenaventura.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'DIR-COMERCIAL': 'DIRECCION',  'BARRIO-COMERCIAL': 'BARRIO',  'MUN-COMERCIAL': 'MUNICIPIO'})


Cartago  = df_Cartago .\
    rename(columns={'ULT-ANO_REN': 'ULT_REN','NIT-PROP.':'NIT', 'DIR-COMERCIAL': 'DIRECCION',  'BARRIO-COMERCIAL': 'BARRIO', 'MUN-COMERCIAL': 'MUNICIPIO'})


Buga  = df_Buga .\
    rename(columns={'ULT-ANO_REN': 'ULT_REN','NIT-PROP.':'NIT', 'DIR-COMERCIAL': 'DIRECCION', 'MUN-COMERCIAL': 'MUNICIPIO'})



Cali = df_Cali.rename(columns={'NUM_ID': 'NIT', 'DIRECCION_COMERCIAL': 'DIRECCION', 'CIUDAD': 'MUNICIPIO_Clean', 'VAL_ULT_CUOTA': 'VLR-ESTABLEC.'})

Cartago_pyg = df_Cartago_pyg.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'DIR-COMERCIAL': 'DIRECCION',  'BARRIO-COMERCIAL': 'BARRIO', 'MUN-COMERCIAL': 'MUNICIPIO'})


Palmira = df_Palmira.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'DIR-COMERCIAL': 'DIRECCION',  'BARRIO-COMERCIAL': 'BARRIO', 'NIT-PROP.':'NIT', 'CIIU':'CIIU-1', 'MUN-COMERCIAL': 'MUNICIPIO'})


Sevilla = df_Sevilla.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'NIT-PROP.': 'NIT', 'DIR-COMERCIAL': 'DIRECCION',  'MUN-COMERCIAL': 'MUNICIPIO'})


Tulua = df_Tulua.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'DIR-COMERCIAL': 'DIRECCION',  'BARRIO-COMERCIAL': 'BARRIO', 'MUN-COMERCIAL': 'MUNICIPIO'})


Sevilla_pyg = df_Sevilla_pyg.\
    rename(columns={'ULT-ANO_REN': 'ULT_REN', 'DIR-COMERCIAL': 'DIRECCION', 'NIT-PROP.':'NIT', 'MUN-COMERCIAL': 'MUNICIPIO'})

df_Ventas = df_Ventas.rename(columns={'Número Nit': 'NIT'})

```

## 4. Consolidación del Dataframe
## 4. Consolidation of the Dataframe

```python
""" Como particularidad de los archivos suministrados, el municipio de cali al ser capital del departamento cuanta con una
estructura de datos mas amplia que la del resto de municipios, por tanto su analisis y union al dataframe consolidado se 
realiza en dos pasos

# As a particularity of the files provided, the municipality of Cali being the capital of the department , has a data 
structure broader than that of the other municipalities, therefore its analysis and binding to the consolidated dataframe
is done in two steps

"""
######## Big Dataframe all columns sin Cali ##########
Municipios_sinCali = pd.concat([Buenaventura, Cartago, Buga, Cartago_pyg, Palmira, Sevilla, Tulua, Sevilla_pyg], sort=False)
Municipios_sinCali[["COD_MUNICIPIO","MUNICIPIO_Clean"]] = Municipios_sinCali["MUNICIPIO"].astype("str").str.split(pat="-",expand=True)
Municipios_sinCali["MUNICIPIO_Clean"] = Municipios_sinCali["MUNICIPIO_Clean"].astype("str").str.strip().str.lower().str.capitalize()

######## all economic activities and sells ##########
Actividades_Economicas = pd.concat([df_Licores, df_Contrib_deporte, df_Loterias], sort=False)
#Actividades_Economicas = pd.merge(df_Licores, df_Ventas,  how = 'outer', left_on ='NIT', right_on = 'Número Nit')
#Actividades_Economicas = pd.merge(Actividades_Economicas, df_Contrib_deporte,  how = 'outer', left_on ='NIT', right_on = 'NIT')
#Actividades_Economicas = pd.merge(Actividades_Economicas, df_Loterias,  how = 'outer', left_on ='NIT', right_on = 'NIT')
```

```python
Municipios_sinCali[["CODIGO_CIIU", 'borrar', "DESC_CIIU"]]=Municipios_sinCali["CIIU-1"].astype("str").str.split(pat="*",expand=True)
Municipios_sinCali["CODIGO_CIIU"]=Municipios_sinCali["CODIGO_CIIU"].astype("str").str.strip()
Municipios_sinCali=Municipios_sinCali.drop(columns=['borrar'])


Municipios_sinCali["CODIGO_CIIU"] = Municipios_sinCali["CODIGO_CIIU"].astype("str")
Municipios_sinCali['CODIGO_CIIU'] = Municipios_sinCali['CODIGO_CIIU'].astype("str").map(lambda x: x.rstrip('aAbBcC'))
#Municipios_sinCali["CODIGO_CIIU"]=[''.join(re.findall("\d*\.?\d+", item)) for item in Municipios_sinCali["CODIGO_CIIU"]]

Municipios = pd.concat([Municipios_sinCali, Cali])
Municipios["CODIGO_CIIU"] = Municipios["CODIGO_CIIU"].astype("str").str.replace(r'\D', '')
Municipios["CODIGO_CIIU"] = Municipios["CODIGO_CIIU"].astype("str")
```

```python
# Se Exportan y guardan los dataframes consolidados para los municipios y actividaddes economicas
# Consolidated dataframes are exported and stored for municipalities and economic activities
Municipios.to_csv('Municipios.csv', index=False)
Actividades_Economicas.to_csv('Actividades_Economicas.csv', index=False)
```
